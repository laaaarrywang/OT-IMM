# Testing Time-Indexed RealNVP Implementation

This notebook comprehensively tests the time-indexed RealNVP flow T(t, x) for:
1. **Invertibility**: T^{-1}(t, T(t, x)) = x
2. **Differentiability w.r.t. time**: ∂T/∂t exists and is smooth
3. **Differentiability w.r.t. input**: ∂T/∂x exists with correct Jacobian
4. **Log-determinant consistency**: log|det(∂T/∂x)| matches forward and inverse
5. **Multiple scales**: From tiny 8×8 to ImageNet 224×224

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from realnvp import TimeIndexedRealNVP, create_vector_flow, create_cifar10_flow, create_imagenet_flow
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
print("="*80)
```

## 0. Check Residual Blocks

```python
def check_residual_blocks(model):
    """Check if model contains residual blocks and count parameters."""
    has_residual = False
    has_se = False
    coupling_params = []
    
    for name, module in model.named_modules():
        if 'ResidualBlock2d' in str(type(module)):
            has_residual = True
        if 'SEBlock' in str(type(module)):
            has_se = True
        if 'ImageAffineCouplingLayer' in str(type(module)):
            # Check for residual-specific attributes
            has_blocks = hasattr(module, 'blocks')
            has_input_conv = hasattr(module, 'input_conv')
            param_count = sum(p.numel() for p in module.parameters())
            coupling_params.append({
                'has_blocks': has_blocks,
                'has_input_conv': has_input_conv,
                'params': param_count
            })
    
    return {
        'has_residual': has_residual,
        'has_se': has_se,
        'coupling_layers': coupling_params,
        'total_params': sum(p.numel() for p in model.parameters())
    }
```

## 1. Test Invertibility

Verify that T^{-1}(t, T(t, x)) = x for various t values and both vector/image inputs.

```python
def test_invertibility(flow, x, t, tolerance=1e-5):
    """Test if flow is invertible: x -> y -> x_reconstructed"""
    # Forward transformation
    y, log_det_forward = flow.forward(x, t)
    
    # Inverse transformation 
    x_reconstructed, log_det_inverse = flow.inverse(y, t)
    
    # Compute reconstruction error
    recon_error = torch.mean(torch.abs(x - x_reconstructed)).item()
    
    # Check log determinant consistency
    log_det_error = torch.mean(torch.abs(log_det_forward + log_det_inverse)).item()
    
    return {
        'reconstruction_error': recon_error,
        'log_det_error': log_det_error,
        'success': recon_error < tolerance and log_det_error < tolerance
    }
```

## 2. Test Differentiability w.r.t. Time

Verify that ∂T/∂t exists and compare finite differences with autograd.

```python
def test_time_differentiability(flow, x, t_center=0.5, eps=1e-4):
    """
    Test differentiability w.r.t. time using both finite differences and autograd.
    Both methods compute the same quantity: ∂T/∂t for each output element.
    """
    batch_size = x.shape[0]
    device = x.device
    
    # Method 1: Finite differences (central difference)
    t_plus = torch.full((batch_size,), min(t_center + eps, 1.0), device=device)
    t_minus = torch.full((batch_size,), max(t_center - eps, 0.0), device=device)
    
    with torch.no_grad():
        y_plus, _ = flow.forward(x, t_plus)
        y_minus, _ = flow.forward(x, t_minus)
    
    # Central difference approximation
    dTdt_finite = (y_plus - y_minus) / (2 * eps)
    
    # Method 2: Autograd (exact derivative)
    t = torch.full((batch_size,), t_center, device=device, requires_grad=True)
    y, _ = flow.forward(x, t)
    
    # Compute gradient for each output element w.r.t. its corresponding t
    # This is more efficient than looping over all elements
    dTdt_autograd = torch.zeros_like(y)
    
    # For each sample in batch
    for i in range(batch_size):
        # Create a ones tensor matching the shape of y[i] for grad_outputs
        grad_output = torch.zeros_like(y)
        grad_output[i] = torch.ones_like(y[i])
        
        # Compute gradient: ∂y[i]/∂t[i]
        grad = torch.autograd.grad(
            y, t,
            grad_outputs=grad_output,
            retain_graph=True,
            create_graph=False
        )[0]
        
        # grad[i] contains ∂y[i]/∂t[i]
        # Expand to match y[i] shape and store
        if y.dim() == 4:  # Image case
            dTdt_autograd[i] = grad[i].view(1, 1, 1)
        else:  # Vector case
            dTdt_autograd[i] = grad[i].unsqueeze(-1)
    
    # Compute norms for comparison
    finite_norm = torch.norm(dTdt_finite.view(batch_size, -1), dim=1)
    autograd_norm = torch.norm(dTdt_autograd.view(batch_size, -1), dim=1)
    
    # Compute element-wise relative error
    diff = torch.abs(dTdt_finite - dTdt_autograd)
    relative_error = diff / (torch.abs(dTdt_finite) + 1e-8)
    
    # Check that gradients exist and are finite
    grad_exists = not (torch.isnan(dTdt_autograd).any() or torch.isinf(dTdt_autograd).any())
    
    return {
        'gradient_exists': grad_exists,
        'finite_diff_norm': finite_norm.mean().item(),
        'autograd_norm': autograd_norm.mean().item(),
        'max_relative_error': relative_error.max().item(),
        'mean_relative_error': relative_error.mean().item(),
        'finite_diff_std': finite_norm.std().item(),
        'autograd_std': autograd_norm.std().item(),
    }

```

## 3. Test Differentiability w.r.t. Input

Verify Jacobian computation and check that transformations are smooth.

```python
def test_time_differentiability(flow, x, t_center=0.5):
    batch_size = x.shape[0]
    t = torch.full((batch_size,), t_center, device=x.device, requires_grad=True)
    
    # Method 1: Finite differences (per-element)
    eps = 1e-4
    t_plus = torch.full((batch_size,), t_center + eps, device=x.device)
    t_minus = torch.full((batch_size,), t_center - eps, device=x.device)
    
    y_plus, _ = flow.forward(x, t_plus)
    y_minus, _ = flow.forward(x, t_minus)
    
    dTdt_finite = (y_plus - y_minus) / (2 * eps)
    
    # Method 2: Autograd (per-element, not sum)
    y, _ = flow.forward(x, t)
    
    # Compute gradient for EACH output element
    dTdt_autograd = torch.zeros_like(y)
    for i in range(batch_size):
        for j in range(y[i].numel()):
            y_flat = y[i].flatten()
            grad = torch.autograd.grad(
                y_flat[j], t[i],
                retain_graph=True,
                create_graph=False
            )[0]
            dTdt_autograd[i].flatten()[j] = grad
    
    # Now both compute the same thing
    finite_norm = torch.norm(dTdt_finite.view(batch_size, -1), dim=1).mean().item()
    autograd_norm = torch.norm(dTdt_autograd.view(batch_size, -1), dim=1).mean().item()
    
    return {
        'dTdt_norm_finite': finite_norm,
        'dTdt_norm_autograd': autograd_norm,
        'relative_error': abs(finite_norm - autograd_norm) / finite_norm
    }
```

## 4. Test Continuous Trajectory

Verify that trajectories T(t, x₀) are continuous and smooth in time.

```python
def test_trajectory_continuity(flow, x0, num_steps=50):
    """
    Test that trajectories are continuous by checking smoothness.
    """
    # Sample trajectory
    trajectory, t_values = flow.sample_trajectory(x0, num_steps=num_steps)
    
    # Compute differences between consecutive points
    diffs = torch.diff(trajectory, dim=0)
    step_sizes = torch.norm(diffs.view(diffs.shape[0], -1), dim=1)
    
    # Check for smoothness (no sudden jumps)
    max_step = step_sizes.max().item()
    mean_step = step_sizes.mean().item()
    std_step = step_sizes.std().item()
    
    # Trajectory should be smooth (small variance in step sizes)
    is_smooth = std_step / mean_step < 0.5  # Coefficient of variation < 0.5
    
    return {
        'max_step': max_step,
        'mean_step': mean_step,
        'std_step': std_step,
        'cv': std_step / mean_step,
        'is_smooth': is_smooth
    }
```

```python
print("\nTesting Input Differentiability")
print("="*50)

# Test with small dimension for detailed analysis
print("1. Vector Mode - Small Dimension (d=16)")
print("-"*40)
small_flow = create_vector_flow(16, num_layers=4).to(device)
x_small = torch.randn(1, 16, device=device)
t_small = torch.tensor([0.5], device=device)

result = test_input_differentiability(small_flow, x_small, t_small)
print("  Gradient exists: {}".format(result['gradient_exists']))
print("  Log-det finite: {}".format(result['log_det_finite']))
print("  Trace finite: {}".format(result['trace_finite']))
print("  ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
print("  log|det(J)|: {:.4f}".format(result['log_det_mean']))
print("  trace(J) estimate: {:.4f}".format(result['trace_mean']))

# Test with larger dimension
print("\n2. Vector Mode - Large Dimension (d={})".format(dim))
print("-"*40)
x_large = torch.randn(4, dim, device=device)
t_large = torch.full((4,), 0.5, device=device)

result = test_input_differentiability(flow_vector, x_large, t_large)
print("  Gradient exists: {}".format(result['gradient_exists']))
print("  Log-det finite: {}".format(result['log_det_finite']))
print("  ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
print("  log|det(J)|: {:.4f}".format(result['log_det_mean']))

# Test image mode without residuals
print("\n3. Image Mode - Original (No Residuals)")
print("-"*40)
flow_img_original = TimeIndexedRealNVP(
    shape=(3, 8, 8),
    num_layers=2,
    hidden_channels=32,
    time_embed_dim=64,
    conv_num_blocks=3,  # Ignored in original
    use_conv_batch_norm=False,
    use_se_blocks=False
).to(device)

x_img = torch.randn(2, 3, 8, 8, device=device)
t_img = torch.full((2,), 0.5, device=device)

result = test_input_differentiability(flow_img_original, x_img, t_img)
print("  Image shape: 3×8×8")
print("  Gradient exists: {}".format(result['gradient_exists']))
print("  Log-det finite: {}".format(result['log_det_finite']))
print("  ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
print("  log|det(J)|: {:.4f}".format(result['log_det_mean']))

# Test image mode with residuals
print("\n4. Image Mode - With Residual Blocks")
print("-"*40)
flow_img_residual = TimeIndexedRealNVP(
    shape=(3, 8, 8),
    num_layers=2,
    hidden_channels=32,
    time_embed_dim=64,
    conv_num_blocks=3,  # Now used!
    use_conv_batch_norm=True,
    use_se_blocks=False
).to(device)

result = test_input_differentiability(flow_img_residual, x_img, t_img)
print("  Image shape: 3×8×8")
print("  Gradient exists: {}".format(result['gradient_exists']))
print("  Log-det finite: {}".format(result['log_det_finite']))
print("  ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
print("  log|det(J)|: {:.4f}".format(result['log_det_mean']))

# Test with SE blocks
print("\n5. Image Mode - With Residual + SE Blocks")
print("-"*40)
flow_img_se = TimeIndexedRealNVP(
    shape=(3, 8, 8),
    num_layers=2,
    hidden_channels=32,
    time_embed_dim=64,
    conv_num_blocks=3,
    use_conv_batch_norm=True,
    use_se_blocks=True  # Enable SE blocks
).to(device)

result = test_input_differentiability(flow_img_se, x_img, t_img)
print("  Image shape: 3×8×8")
print("  Gradient exists: {}".format(result['gradient_exists']))
print("  Log-det finite: {}".format(result['log_det_finite']))
print("  ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
print("  log|det(J)|: {:.4f}".format(result['log_det_mean']))

# Test CIFAR-10 scale with residuals
print("\n6. CIFAR-10 Scale - Comparing Gradient Properties")
print("-"*40)
x_cifar = torch.randn(2, 3, 32, 32, device=device)
t_cifar = torch.full((2,), 0.5, device=device)

configs = [
    ("Original", False, False),
    ("With Residuals", True, False),
    ("With Residuals + SE", True, True)
]

for name, use_bn, use_se in configs:
    flow_cifar = TimeIndexedRealNVP(
        shape=(3, 32, 32),
        num_layers=2,  # Small for testing
        hidden_channels=64,
        time_embed_dim=128,
        conv_num_blocks=3,
        use_conv_batch_norm=use_bn,
        use_se_blocks=use_se
    ).to(device)
    
    result = test_input_differentiability(flow_cifar, x_cifar, t_cifar)
    print("\n  {}:".format(name))
    print("    Gradient exists: {}".format(result['gradient_exists']))
    print("    Log-det finite: {}".format(result['log_det_finite']))
    print("    ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
    print("    log|det(J)|: {:.4f}".format(result['log_det_mean']))

# Test gradient stability over time
print("\n7. Gradient Stability Over Time (With Residuals)")
print("-"*40)
flow_test = create_cifar10_flow(
    num_layers=2,
    hidden_channels=64,
    conv_num_blocks=3,
    use_conv_batch_norm=True,
    use_se_blocks=False
).to(device)

x_test = torch.randn(2, 3, 32, 32, device=device)
time_points = [0.0, 0.25, 0.5, 0.75, 1.0]
grad_norms = []

print("  Time   ||∂(sum T)/∂x||   log|det(J)|")
print("  " + "-"*36)
for t_val in time_points:
    t_test = torch.full((2,), t_val, device=device)
    result = test_input_differentiability(flow_test, x_test, t_test)
    grad_norms.append(result['grad_norm'])
    print("  {:.2f}    {:>12.4f}    {:>11.4f}".format(
        t_val, result['grad_norm'], result['log_det_mean']))

# Check gradient norm variation
grad_std = torch.std(torch.tensor(grad_norms)).item()
print("\n  Gradient norm std dev: {:.4f}".format(grad_std))
print("  Gradient stability: {}".format(
    "Good" if grad_std < 10.0 else "May need adjustment"))

# Test with very deep network (stress test)
print("\n8. Deep Network Stress Test")
print("-"*40)
try:
    deep_configs = [
        ("Deep Original", 8, False, False),
        ("Deep with Residuals", 8, True, False),
        ("Very Deep with Residuals", 12, True, False)
    ]
    
    for name, layers, use_bn, use_se in deep_configs:
        flow_deep = TimeIndexedRealNVP(
            shape=(3, 16, 16),  # Smaller image for memory
            num_layers=layers,
            hidden_channels=32,
            time_embed_dim=64,
            conv_num_blocks=3,
            use_conv_batch_norm=use_bn,
            use_se_blocks=use_se
        ).to(device)
        
        x_deep = torch.randn(1, 3, 16, 16, device=device)
        t_deep = torch.tensor([0.5], device=device)
        
        result = test_input_differentiability(flow_deep, x_deep, t_deep)
        print("\n  {} ({} layers):".format(name, layers))
        print("    Gradient exists: {}".format(result['gradient_exists']))
        print("    ||∂(sum T)/∂x||: {:.4f}".format(result['grad_norm']))
        print("    log|det(J)|: {:.4f}".format(result['log_det_mean']))
        
        # Check for vanishing/exploding gradients
        if result['grad_norm'] < 0.01:
            print("    ⚠️  Warning: Possible vanishing gradients")
        elif result['grad_norm'] > 1000:
            print("    ⚠️  Warning: Possible exploding gradients")
        else:
            print("    ✓ Gradient magnitude healthy")
            
except Exception as e:
    print("  Error in deep network test: {}".format(str(e)))

print("\n" + "="*50)
```

```python
import time

def benchmark_performance():
    """
    Benchmark forward and inverse pass times for different architectures.
    """
    print("\nPERFORMANCE BENCHMARK")
    print("="*50)
    
    # Test configurations comparing original vs residual
    print("\n1. Architecture Comparison on Same Input Size")
    print("-"*40)
    
    # Small image benchmark
    shape = (10, 3, 16, 16)
    x = torch.randn(shape, device=device)
    t = torch.rand(shape[0], device=device)
    
    arch_configs = [
        ("Original", False, False, False),
        ("With Residuals", True, True, False),
        ("With Residuals + SE", True, True, True)
    ]
    
    for name, use_bn, use_res, use_se in arch_configs:
        try:
            flow = TimeIndexedRealNVP(
                shape=(3, 16, 16),
                num_layers=4,
                hidden_channels=32,
                time_embed_dim=64,
                conv_num_blocks=3,
                use_conv_batch_norm=use_bn,
                use_se_blocks=use_se
                use_residual = use_res
            ).to(device)
            
            # Warmup
            for _ in range(3):
                _ = flow.forward(x, t)
            
            # Benchmark forward pass
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            num_runs = 20
            for _ in range(num_runs):
                y, log_det = flow.forward(x, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            forward_time = (time.time() - start) / num_runs * 1000  # ms
            
            # Benchmark inverse pass
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            for _ in range(num_runs):
                x_rec, log_det_inv = flow.inverse(y, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            inverse_time = (time.time() - start) / num_runs * 1000  # ms
            
            print(f"\n  {name}:")
            print(f"    Forward: {forward_time:.2f} ms")
            print(f"    Inverse: {inverse_time:.2f} ms")
            print(f"    Params: {sum(p.numel() for p in flow.parameters()):,}")
            
        except Exception as e:
            print(f"\n  {name}: Error - {str(e)[:50]}")
    
    # General benchmark across different sizes
    print("\n2. Performance Across Different Input Sizes")
    print("-"*40)
    
    configs = [
        ("Small Vector (d=32)", create_vector_flow(32, num_layers=4), (100, 32)),
        ("Large Vector (d=256)", create_vector_flow(256, num_layers=8), (100, 256)),
        ("Single Channel (1×16×16)", TimeIndexedRealNVP((1, 16, 16), num_layers=4, conv_num_blocks=3, use_conv_batch_norm=True), (100, 1, 16, 16)),
        ("RGB Image (3×16×16)", TimeIndexedRealNVP((3, 16, 16), num_layers=4, conv_num_blocks=3, use_conv_batch_norm=True), (100, 3, 16, 16)),
    ]
    
    for name, flow, shape in configs:
        try:
            flow = flow.to(device)
            x = torch.randn(shape, device=device)
            t = torch.rand(shape[0], device=device)
            
            # Warmup
            for _ in range(5):
                _ = flow.forward(x, t)
            
            # Benchmark forward pass
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            num_runs = 20
            for _ in range(num_runs):
                y, log_det = flow.forward(x, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            forward_time = (time.time() - start) / num_runs * 1000  # ms
            
            # Benchmark inverse pass
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            for _ in range(num_runs):
                x_rec, log_det_inv = flow.inverse(y, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            inverse_time = (time.time() - start) / num_runs * 1000  # ms
            
            print(f"\n  {name}:")
            print(f"    Shape: {shape}")
            print(f"    Forward: {forward_time:.2f} ms")
            print(f"    Inverse: {inverse_time:.2f} ms")
            print(f"    Params: {sum(p.numel() for p in flow.parameters()):,}")
            
        except Exception as e:
            print(f"\n  {name}: Error - {str(e)[:50]}")
    
    # Benchmark specifically for CIFAR-10 and ImageNet scales with residuals
    print("\n3. Large-Scale Benchmarks (with Residuals)")
    print("-"*40)
    
    large_configs = [
        ("CIFAR-10 Original", create_cifar10_flow(num_layers=2, hidden_channels=64, use_conv_batch_norm=False), (10, 3, 32, 32)),
        ("CIFAR-10 w/ Residuals", create_cifar10_flow(num_layers=2, hidden_channels=64, conv_num_blocks=3, use_conv_batch_norm=True), (10, 3, 32, 32)),
        ("ImageNet-64 Original", TimeIndexedRealNVP((3, 64, 64), num_layers=2, hidden_channels=128, use_conv_batch_norm=False), (2, 3, 64, 64)),
        ("ImageNet-64 w/ Residuals", TimeIndexedRealNVP((3, 64, 64), num_layers=2, hidden_channels=128, conv_num_blocks=3, use_conv_batch_norm=True, use_se_blocks=True), (2, 3, 64, 64)),
    ]
    
    for name, flow, shape in large_configs:
        try:
            flow = flow.to(device)
            x = torch.randn(shape, device=device)
            t = torch.rand(shape[0], device=device)
            
            # Warmup
            for _ in range(3):
                _ = flow.forward(x, t)
            
            # Benchmark with fewer runs for large models
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            num_runs = 10
            for _ in range(num_runs):
                y, log_det = flow.forward(x, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            forward_time = (time.time() - start) / num_runs * 1000  # ms
            
            # Inverse pass
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            
            for _ in range(num_runs):
                x_rec, log_det_inv = flow.inverse(y, t)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            inverse_time = (time.time() - start) / num_runs * 1000  # ms
            
            print(f"\n  {name}:")
            print(f"    Forward: {forward_time:.2f} ms")
            print(f"    Inverse: {inverse_time:.2f} ms")
            print(f"    Params: {sum(p.numel() for p in flow.parameters()):,}")
            print(f"    Memory: ~{sum(p.numel() for p in flow.parameters()) * 4 / 1024**2:.1f} MB")
            
        except Exception as e:
            print(f"\n  {name}: Error - {str(e)[:50]}")
    
    # Performance/parameter trade-off analysis
    print("\n4. Performance/Parameter Trade-off Analysis")
    print("-"*40)
    print("\nExpected trade-offs with residual blocks:")
    print("  • ~2x more parameters")
    print("  • ~1.5-2x slower inference (more operations)")
    print("  • Better gradient flow during training")
    print("  • More stable optimization for deep networks")
    print("  • Higher memory usage")

benchmark_performance()

print("\n" + "="*50)
```
